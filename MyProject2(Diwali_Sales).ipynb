```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Reading the file which is of type csv
df = pd.read_csv('https://raw.githubusercontent.com/rishabhnmishra/Python_Diwali_Sales_Analysis/main/Diwali%20Sales%20Data.csv',encoding = 'unicode_escape')
df
```

```python
#Removing columns with null values
df = df.drop(['Status' , 'unnamed1'], axis = 1)
df
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.info()
```

```python
df['Amount'] = df['Amount'].astype('int64')
```

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
  ax.bar_label(bars)           #There are more females than man
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)   #Females spent more amount than man
```

```python
ax = sns.countplot(data = df, x = 'Age Group' , hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)               #Age wise distribution of people w.r.t their gender
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)   #The people of age group 26-35 spent the most
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).loc[(df['Marital_Status']==1)]
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)     #People who are married and are in age group of 26-35 have spent more than others
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).loc[(df['Marital_Status']==0)]
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(5,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders')
plt.xticks(rotation = 90)           #People from UP have the most orders
```

```python
sales_state = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(5,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Amount')
plt.xticks(rotation = 90)                             #People from UP have spent the most
```

```python
ax = sns.countplot(data = df, x = 'Marital_Status')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_marital = df.groupby(['Marital_Status' , 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(data = sales_marital, x = 'Marital_Status', y = 'Amount', hue = 'Gender')
```

```python
sns.set(rc = {'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_occ = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(data = sales_occ, x = 'Occupation', y = 'Amount')
```

```python
ax = sns.countplot(data = df,x = 'Product_Category')
for bars in ax.containers:
  ax.bar_label(bars)
plt.xticks(rotation = 90)
```

```python
sales_pro = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.barplot(data = sales_pro, x = 'Product_Category', y = 'Amount')
plt.xticks(rotation = 45)
```

```python
sales_id = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.barplot(data = sales_id, x = 'Product_ID', y = 'Orders')
```
